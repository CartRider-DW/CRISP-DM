### 상호명 과 지점명 병합

```python
# 컬럼 병합 함수
def combine_2rd_columns(col_1, col_2):
    result = col_1
    if not pd.isna(col_2):
        result += " " + str(col_2)
    return result

df2["상호-지점명"] = df2.apply(lambda x: combine_2rd_columns(x['상호명'], x['지점명']), axis=1)
```
