### 데이터 정제
#### 최종본세, 부과번호, 최종납기일자 drop
- 관련 조문
    -  질서위반행위규제법 제18조(자진납부자에 대한 과태료 감경) ① 행정청은 당사자가 제16조에 따른 의견 제출 기한 이내에 과태료를 자진하여 납부하고자 하는 경우에는 대통령령으로 정하는 바에 따라 과태료를 감경할 수 있다.
    - 질서위반행위규제법 제24조(가산금 징수 및 체납처분 등) ① 행정청은 당사자가 납부기한까지 과태료를 납부하지 아니한 때에는 납부기한을 경과한 날부터 체납된 과태료에 대하여 100분의 3에 상당하는 가산금을 징수한다. <개정 2016. 12. 2.> ② 체납된 과태료를 납부하지 아니한 때에는 납부기한이 경과한 날부터 매 1개월이 경과할 때마다 체납된 과태료의 1천분의 12에 상당하는 가산금(이하 이 조에서 “중가산금”이라 한다)을 제1항에 따른 가산금에 가산하여 징수한다. 이 경우 중가산금을 가산하여 징수하는 기간은 60개월을 초과하지 못한다. ③ 행정청은 당사자가 제20조제1항에 따른 기한 이내에 이의를 제기하지 아니하고 제1항에 따른 가산금을 납부하지 아니한 때에는 국세 또는 지방세 체납처분의 예에 따라 징수한다.

#### 부과대상이 대전인 행만 남김

```python
df_daejeon = df[df['부과대상'].str.contains('대전', na=False)]
df_daejeon
```

#### 부과연도, 부과월, 부과일 분리

```python
list = df_daejeon['부과일자'].to_list()
list_year = []
for i in list:
    list_year.append(i.split('-')[0])
list_year
```

```python
list_month = []
for i in list:
    list_month.append(i.split('-')[1])
list_month
```

```python
list_day = []
for i in list:
    list_day.append(i.split('-')[2])
list_day
```

```python
df_daejeon['부과연도'] = list_year
df_daejeon['부과월'] = list_month
df_daejeon['부과일'] = list_day
```

```python
df_daejeon
```

```python
# 부과연도 및 부과월별 과태료 부과건수
df_pivot = pd.pivot_table(df_daejeon,
                          index = ['부과연도'],
                          columns = '부과월',
                          values = '부과대상',
                          aggfunc = 'count',
                          margins = True
                          )

df_pivot
```

#### 평일/주말 여부 구분

```python
from datetime import datetime

date_list =[]
for i in df_daejeon['부과일자']:
    date = datetime.strptime(i, '%Y-%m-%d')
    date_num = date.weekday()
    date_list.append(date_num)

    # 0은 월요일, 1은 화요일, 2는 수요일, 3은 목요일, 4는 금요일 ,5는 토요일 ,6은 일요일

print(date_list)
```

```python
date_list_final=[]
for i in date_list:
    if i <= 4:
        date_list_final.append('평일')
    elif 5<= i<= 6:
        date_list_final.append('주말')

print(date_list_final)
```

```python
Counter(date_list_final)
# 평일에 부과된 경우가 압도적으로 많았음
# 과태료 부과업무가 평일에 이루어지기 때문인 것으로 파악됨
# 따라서 분석에 유의미하지 않다고 보아 별도의 열 분리는 하지 않음.
```
