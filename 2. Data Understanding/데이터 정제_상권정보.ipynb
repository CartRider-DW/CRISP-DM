### 4-3. feature 결합

```python
comm['지점명'].fillna('', inplace = True)
    # NaN --> 공란 변경, 저장

comm['상호+지점명'] = comm['상호명'] + ' ' + comm['지점명']
    # 상호명 + 지점명 결합 --> 새로운 열, 상호+지점명 생성
```

```python
comm.drop(['상호명', '지점명'], axis = 1, inplace = True)
```
