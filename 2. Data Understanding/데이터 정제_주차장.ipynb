### 데이터 정제

```python
#부제시행구분: 차량 2부제, 5부제 등 = 차량운행을 줄이기 위한 제도
# 시 내의 차량 통행량을 줄이기 위한 제도는 흡연 금지구역과는 무관함
print(df.shape)
df = df.drop('부제시행구분', axis = 1)
print(df.shape)
```

```python
#추가단위시간: 주차를 기본 시간 이상으로 했을 때, 어떤 단위로 계산하는지
print(df.shape)
df = df.drop('추가단위시간', axis = 1)
print(df.shape)
```

```python
#추가단위요금: 주차를 기본 시간 이상으로 했을 때, 추가 금액
print(df.shape)
df = df.drop('추가단위요금', axis = 1)
print(df.shape)
```

```python
#1일주차권요금적용시간: 각 주차장으로 1일 주차권 요금 적용시간
print(df.shape)
df = df.drop('1일주차권요금적용시간', axis = 1)
print(df.shape)
```

```python
#1일주차권요금: 주차 요금은 흡연과 무관, 더 많은 사람들이 이용할 수 있지만
#               저렴해서 이용량을 증가하고 그 영향이 흡연에도 영향을 미치기 힘듬
print(df.shape)
df = df.drop('1일주차권요금', axis = 1)
print(df.shape)
```

```python
#월정기권요금
print(df.shape)
df = df.drop('월정기권요금', axis = 1)
print(df.shape)
```

```python
#결제방법: 흡연과 무관
print(df.shape)
df = df.drop('결제방법', axis = 1)
print(df.shape)
```

```python
#특기사항: 결측치가 상당히 많아서 전반적 데이터 분석에 도움이 안됨
print(df.shape)
df = df.drop('특기사항', axis = 1)
print(df.shape)
```

```python
#전화번호
print(df.shape)
df = df.drop('전화번호', axis = 1)
print(df.shape)
```

```python
#제공기관코드: 데이터를 제공한 기관은 주차장 위치랑 크게 상관이 없음
print(df.shape)
df = df.drop('제공기관코드', axis = 1)
print(df.shape)
```

```python
#제공기관명
print(df.shape)
df = df.drop('제공기관명', axis = 1)
print(df.shape)
```

```python
#unnamed: 대부분의 데이터가 결측치이며, 어떤 이유로 생긴 컬럼인지 구별하기 힘듬
print(df.shape)
df = df.dropna(thresh=10000, axis = 1)
print(df.shape)
```
