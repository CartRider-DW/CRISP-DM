```python
import pandas as pd
pd.set_option("display.max_rows", 100, "display.max_columns", 100)
```

```python
PATH = 'D:/project/데이터 분석 프로젝트/이노포스트/data/'

df = pd.read_csv(PATH + '06_주차장/전국주차장정보표준데이터.csv', encoding='cp949')
```

```python
# 대전 지역의 데이터만 추출, `소재지지번주소`에서 `대전`이 포함된 데이터 추출
df2 = df[df['소재지지번주소'].str.contains('대전', na=False)]
df2.head(2)
```

```python
df2.shape
```

```python
# `소재지지번주소`의 NaN 중 `대전` 지역의 데이터 추출, NaN 중 `소재지도로명주소`에서 `대전`인 값
df3 = df[df['소재지지번주소'].isnull()] # 소재지지번주소에서 NaN 추출
df4 = df3[df3['소재지도로명주소'].str.contains('대전', na=False)] # df3에서 소재지도로명주소에 대전이 있는 값 추출
df4.head() # 2건 있음
```

```python
df5 = pd.concat([df2,df4], ignore_index=True)
df5.shape
```

```python
df5.head()
```

```python
df5.info()
```
